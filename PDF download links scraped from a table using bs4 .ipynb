```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup


url = 'https://www.4c-services.org/process/list-of-4c-certificates/'
data  = requests.get(url).text
soup = BeautifulSoup(data)

links = BeautifulSoup(data).select('td a')
urls = [link['href'] for link in links]
print(urls)

pdf = pd.DataFrame (urls, columns = ['PDF'])
print(pdf)
```
