```python
import selenium                                                
import pandas as pd 
import re
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains
import string
import os, platform
import requests
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager      
driver = webdriver.Chrome(ChromeDriverManager().install())
#driver = webdriver.Chrome(executable_path='/Users/User/Upwork Project/PlanetX/chromedriver')
driver.maximize_window()
```

```python
url = 'https://www.smartwatermark.org/products/waterwise/?page=1'
driver.get(url)
```

```python
#input 4
pages = input('How many pages? ')
pages = int(pages)
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0], 
                columns=['Product Name',
                        'Product Description',
                        'Benefits',
                        'More Information Link'])
gc
```

```python
t=2
for k in range(pages):
    cc = driver.find_elements(By.CSS_SELECTOR, 'body > div.page-wrap > div.container > div.row.row-flex.row-flex-wrap.thumbnail-group > div.col-xs-6.col-sm-4.col-md-3 > div > a > div.caption > h2')
    cc = len(cc)
    cc = int(cc)
    for i in range(cc):
        product = driver.find_elements(By.CSS_SELECTOR, 'body > div.page-wrap > div.container > div.row.row-flex.row-flex-wrap.thumbnail-group > div.col-xs-6.col-sm-4.col-md-3 > div > a > div.caption > h2')
        product = product[i].text
        lnk = driver.find_elements(By.CSS_SELECTOR, 'body > div.page-wrap > div.container > div.row.row-flex.row-flex-wrap.thumbnail-group > div.col-xs-6.col-sm-4.col-md-3 > div > a')
        lnk = lnk[i].get_attribute('href')
        
        # Switch to the New window
        driver.execute_script("window.open('');")
        driver.switch_to.window(driver.window_handles[1])
        driver.get(lnk)
        time.sleep(7)
        
        try:
            bn = driver.find_element(By.CSS_SELECTOR, 'body > div.page-wrap > div.container > div.row > div > ul')
            bn = bn.text
        except:
            bn = 'none'
            
            
        try:
            des = driver.find_element(By.CSS_SELECTOR, 'body > div.page-wrap > div.container > div.row > div > p')
            des = des.text
        except:
            des = 'none'
            
            
        try:
            mil = driver.find_element(By.CSS_SELECTOR, 'body > div.page-wrap > div.container > div.item-button-list > a:nth-child(1)')
            mil = mil.get_attribute('href')
            if mil == '/products/':
                mil = 'None'
            else: 
                mil = mil
        except:
            mil = 'None'
            
        Dic = {'Product Name': product,
                        'Product Description': des,
                        'Benefits': bn,
                        'More Information Link' : mil}    
        gc = gc.append(Dic, ignore_index=True)
        # Switch to the Previous window
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    # Go to the Next Page
    base = 'https://www.smartwatermark.org/products/waterwise/'
    extra = '?page='
    n = str(t)
    url = base+extra+n
    driver.get(url)
    time.sleep(5)
    t = int(t)
    t = t+1
```
