```python
import selenium                                                
import pandas as pd 
import re
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.common.action_chains import ActionChains
import string
import os, platform
import requests
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager      
driver = webdriver.Chrome(ChromeDriverManager().install())
#driver = webdriver.Chrome(executable_path='/Users/User/Upwork Project/PlanetX/chromedriver')
driver.maximize_window()
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0], 
                columns=['Hospital Name',
                        'Network',
                        'Address',
                        'Map Link',
                        'Phone',
                        'Website'])
gc
```

```python
url = 'https://www.hrc.org/resources/healthcare-facilities/search/p48?designation=leader'
driver.get(url)
```

```python
#input 3
pages = input('How many pages? ')
pages = int(pages)
```

```python
t = 49
for k in range(pages):
    ccc = driver.find_elements(By.CSS_SELECTOR, '#main > div > section > div > article > div > div > h2 > a')
    ccc = len(ccc)
    ccc = int(ccc)
    for i in range(ccc):
        webb = driver.find_elements(By.CSS_SELECTOR, '#main > div > section > div > article > div > div > h2 > a')
        webb = webb[i].get_attribute('href')
        
        
        # Switch to the New window
        driver.execute_script("window.open('');")
        driver.switch_to.window(driver.window_handles[1])
        driver.get(webb)
        time.sleep(5)
        
        hnam = driver.find_element(By.CSS_SELECTOR,'#main > header.section.bg-blue-400.text-white > div > div > div > div > div > h1')
        hnam = hnam.text
        print(hnam)
        
        try:
            mapl = driver.find_element(By.CSS_SELECTOR, '#main > header.section.bg-blue-400.text-white > div > div > div.flex-grid-col.overflow-hidden.relative.w-full.lg\:w-3\/4 > div > div > dl > div:nth-child(1) > dd.mb-32 > a')
            mapl = mapl.get_attribute('href')
            print(mapl)
        except:
            mapl = 'N/A'
            print(mapl)
                                       
                                  
                                   
        #CODE FOR NETWORK & ADDRESS
        
        for p in range(2):
            try:
                try:
                    cc = driver.find_elements(By.CSS_SELECTOR, '#main > header.section.bg-blue-400.text-white > div > div > div.flex-grid-col.overflow-hidden.relative.w-full.lg\:w-3\/4 > div > div > dl > div:nth-child(1) > dt.type-eyebrow.mb-12')
                    cc = cc[p].text
                    if cc == 'NETWORK':
                        net = driver.find_elements(By.CSS_SELECTOR, '#main > header.section.bg-blue-400.text-white > div > div > div.flex-grid-col.overflow-hidden.relative.w-full.lg\:w-3\/4 > div > div > dl > div:nth-child(1) > dd.mb-32')
                        net = net[p].text
                        n = net
                        print('Network:- ',n)
                        print('value:-',p)
                    else:
                        net = driver.find_elements(By.CSS_SELECTOR, '#main > header.section.bg-blue-400.text-white > div > div > div.flex-grid-col.overflow-hidden.relative.w-full.lg\:w-3\/4 > div > div > dl > div:nth-child(1) > dd.mb-32')
                        net = net[p].text
                        pattern = re.compile(r'((.*(\n|\r|\r\n)){3})')
                        matches = pattern.finditer(net)
                        for match in matches:
                            net = match.group()
                        pattern = re.compile(r'((.*(\n|\r|\r\n)){2})')
                        matches = pattern.finditer(net)
                        for match in matches:
                            net = match.group()
                        net = net.strip()
                        ad = net
                        print('Address:- ',ad)
                        print('value:-',p)
                except:
                    if p > 0:
                        n = 'No Data'
                        print('Network1:- ',n)
                    else:
                        ad = 'No Data'
                        print('Address1:- ',ad)
            except:
                n = 'No Data'
                ad = 'No Data'

        print('Last Network:- ',n)
        print('Last Address:- ',ad)
        
                                       
        #CODE FOR WEBSITE & PHONE
                                       
        for q in range(2):
            try:
                try:
                    cc = driver.find_elements(By.CSS_SELECTOR, '#main > header.section.bg-blue-400.text-white > div > div > div.flex-grid-col.overflow-hidden.relative.w-full.lg\:w-3\/4 > div > div > dl > div:nth-child(2) > dd > a')
                    cc = cc[q].get_attribute('href')
                    pattern = re.compile(r'^(tel)?')
                    matches = pattern.finditer(cc)
                    for match in matches:
                        tt = match.group()
                    print(tt)
                    if tt == 'tel':
                        pattern = re.compile(r'\d.+')
                        matches = pattern.finditer(cc)
                        for match in matches:
                            tlp = match.group()
                        print('Phone:- ',tlp)
                    else:
                        web = cc
                        print('Website:- ',web)

                except:
                    cc = driver.find_elements(By.CSS_SELECTOR, '#main > header.section.bg-blue-400.text-white > div > div > div.flex-grid-col.overflow-hidden.relative.w-full.lg\:w-3\/4 > div > div > dl > div:nth-child(2) > dd > a')
                    cc = cc[0].get_attribute('href')
                    pattern = re.compile(r'^(tel)?')
                    matches = pattern.finditer(cc)
                    for match in matches:
                        tt = match.group()
                    print(tt)
                    if tt == 'tel':
                        pattern = re.compile(r'\d.+')
                        matches = pattern.finditer(cc)
                        for match in matches:
                            tlp = match.group()
                        print('Phone:- ',tlp)
                    else:
                        web = cc
                        tlp = 'No Data'
                        print('Website:- ',web)
                        print('Phone:- ',tlp)
            except:
                web = 'No Data'
                tlp = 'No Data'

        print('Last Website:- ',web)
        print('Last Phone:- ',tlp)
        
        Dic={'Hospital Name': hnam,
                    'Network': n,
                    'Address': ad,
                    'Map Link': mapl,
                    'Phone': tlp,
                    'Website': web}
        gc = gc.append(Dic, ignore_index=True)
        print(gc)
                                       
        # Switch to the Previous window
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

    # Go to the Next Page
    base = 'https://www.hrc.org/resources/healthcare-facilities/search/p'
    extra = '?designation=leader'
    n = str(t)
    url = base+n+extra
    driver.get(url)
    time.sleep(3)
    t = int(t)
    t = t+1
```
